#**Creating a SQL DATABASE from json files**


```
import sqlite3 as sq3
import json
import pandas as pd
```

```
with open("some_movies.json") as f:
  data = json.load(f)
```

```
data
```

```
df = pd.json_normalize(data, sep=",")
```

```
df
```

```
list(df)
```

```
movies = df[["id","release_date","belongs_to_collection","budget","revenue","vote_count","vote_average"]].copy()
movies
```

```
movies.info()
```

```
movies.release_date = pd.to_datetime(df['release_date'],errors='coerce')
movies.release_date
```

```
movies.revenue = df.revenue/1000000
movies.budget = df.revenue/1000000
```

```
movies.revenue
```

```
votes = df[["id","vote_count","vote_average","release_date"]].copy()
votes
```

```
genres = pd.json_normalize(data = data, record_path = "genres", meta = "id", record_prefix="genres_")
genres
```

```
prod = pd.json_normalize(data = data, record_path="production_companies", meta="id", record_prefix="company_")
prod
```

```
con = sq3.connect("movies.db")
con
```

```
con.execute("DROP TABLE movies").fetchall()
```

```
movies.info()
```

```
movies.to_sql("movies", con , index=False, if_exists='replace')
```

```
genres.to_sql("genres",con,index=False)
```

```
prod.to_sql("prod",con,index=False)
```

```
votes.to_sql("votes",con,index=False)
```

```
con.execute("SELECT * FROM sqlite_master").fetchall()
```

```
con.execute("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name").fetchall()
```

```
con.close()
```

#**Loading Data from an SQL DATABASE into DataFrames**

```
import pandas as pd
import sqlite3 as sq3
```

```
con = sq3.connect("movies.db")
```

```
con.execute("SELECT * from sqlite_master").fetchall()
```

```
pd.read_sql("SELECT * FROM votes",con).info()
```

```
df = pd.read_sql("SELECT * FROM movies",con, index_col="id",parse_dates="release_date")
df
```

```
df.info()
```

```
genres = pd.read_sql("SELECT * FROM genres",con, index_col="id")
genres
```

```
con.close()
```
